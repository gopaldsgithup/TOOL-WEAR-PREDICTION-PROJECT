```python
from scipy import stats

def two_sample(d1,d2):
  t=0
  f=0
  for i in  range(31):
    sample1=d1.sample(frac=0.03)
    sample2=d2.sample(frac=0.03)
    t_test,p_value=stats.ttest_ind(sample1,sample2)
    if p_value < 0.055:
      f=f+1
    else:
      t=t+1
  if t>f:
    return True
  else:
    return False

#defining function for categories vs categories
def chisqare_cat_vs_cat(d1,d2):
  return True if stats.chi2_contingency(pd.crosstab(d1,d2))[1] < 0.055 else False

def annova_test(d1, d2):
    # Ensure d1 is numeric
    df1[d1] = pd.to_numeric(df1[d1], errors='coerce')  # Convert non-numeric values to NaN

    group = df1[d2].unique()
    data = {i: df1[d1][df1[d2] == i].dropna() for i in group}  # Drop NaN values

    # Filter out empty groups
    valid_data = [values for values in data.values() if len(values) > 0]

    if len(valid_data) < 2:  # ANOVA requires at least two groups
        return False

    f_value, p_value = stats.f_oneway(*valid_data)
    return p_value >= 0.055


final={}
for i in df1.columns:
  final[i]={}
  for j in df1.columns:
    if (i in continues) and (j in continues):
      result=two_sample(df1[i],df1[j])
    elif  (i in continues) and (j in categories):
      result=annova_test(i,j)
    elif (i in categories) and (j in continues):
      result=annova_test(j,i)
    elif (i in categories) and (j in categories):
      result=chisqare_cat_vs_cat(df1[i],df1[j])
    if result:
      final[i][j]=1
    else:
      final[i][j]=0
```

```python
# Reshape for LSTM (samples, time steps, features)
sequence_length = 10  # Define a reasonable sequence length
X_sequences, y1_sequences, y2_sequences, y3_sequences = [], [], [], []
```

```python
for i in range(len(X) - sequence_length):
    X_sequences.append(X[i:i + sequence_length])
    y1_sequences.append(y1[i + sequence_length])
    y2_sequences.append(y2[i + sequence_length])
    y3_sequences.append(y3[i + sequence_length])

X_sequences = np.array(X_sequences)
y1_sequences = np.array(y1_sequences)
y2_sequences = np.array(y2_sequences)
y3_sequences = np.array(y3_sequences)
```
